```python
# Import the necessary modules
import ee
import geemap
import pandas as pd
import numpy as np

# Initialize the Earth Engine module
ee.Initialize()
#ee.Authenticate()
```

```python
lcc = pd.read_excel('meuse/CORINE/land_cover_classes.xlsx')

dataset = ee.FeatureCollection('projects/mohanrahan/assets/Meuse_Catchment_4326')



years = [1990, 2000, 2006, 2012, 2018]

# Load the Corine Land Cover dataset
def get_LC_proportions(dataset, years):
    for year in years:
        
        landcover = ee.Image(f'COPERNICUS/CORINE/V20/100m/{str(year)}')

        # Get the FeatureCollection of geometries
        data_add_LC = landcover.reduceRegions(**{
            'reducer':ee.Reducer.frequencyHistogram(),
            'collection':dataset,
            'scale':1000,
        })
        
        IDS = data_add_LC.aggregate_array('system:index').getInfo()
        print(IDS)
        df = pd.DataFrame(columns=lcc.Value.values)
        df['system_index'] = IDS
        df.set_index('system_index', inplace=True)
        
        for i in IDS:
            proportions = data_add_LC.filterMetadata('system:index', 'equals', ee.String(i)).getInfo()['features'][-1]['properties']['histogram']#['histogram'])
            prop_df = pd.DataFrame(proportions, index = [i])#.set_index('system_index', inplace=True)
            prop_df['system_index'] = str(i)
            prop_df.set_index('system_index', inplace=True)
            prop_df.to_excel(f'meuse/CORINE/lc_{i}_{year}.xlsx')
            

get_LC_proportions(dataset, years)


# Get the area of each geometry in the FeatureCollection
#areas = fc.map(lambda f: ee.Feature(f.geometry()).area())

# Join the areas to the FeatureCollection
#fc = fc.map(lambda f: f.set('area', areas.get(f.id())))

# Get the CLC value for each geometry
#clc = clc.reduceRegions(collection=fc, reducer=ee.Reducer.first())

# Join the CLC value to the FeatureCollection
#fc = fc.map(lambda f: f.set('clc', clc.filter(ee.Filter.eq('id', f.id())).first().get('first')))

# Get the proportion of each CLC class
#clc_classes = ee.Dictionary(clc.aggregate_histogram('first'))

#clc_classes = clc_classes.map(lambda clc, count: {'clc': clc, 'area': ee.Number(count).divide(areas.size()).multiply(100)})

# Plot the proportions of CLC classes
#from plotly.subplots import make_subplots
#import plotly.graph_objs as go

#fig = make_subplots(rows=1, cols=1)

#data = [go.Bar(x=(clc_classes), y=clc_classes.values().map(lambda x: x.get('area')), name="CLC Classes")]

#fig.add_trace(data[0])
#fig.update_layout(title_text="Corine Land Cover Classes Distribution", xaxis_title="CLC Classes", yaxis_title="Area (%)")
#fig.show()
```
